```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet('../data/datos-todas-estaciones.parquet')
df.sample(4)
```

```python
ids_estaciones = df.id_estacion.unique()
```

```python
def normalizar_nombres_columnas(df):
    return [columna.lower().strip() for columna in df.columns]

def formato_fecha(array):
    return pd.to_datetime(array)
```

```python
df.fecha = formato_fecha(df.fecha)
```

```python
id = ids_estaciones[5]

df.columns = normalizar_nombres_columnas(df)

mini = df.groupby('id_estacion').get_group(id).sort_values('fecha')
```
