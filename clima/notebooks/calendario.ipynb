```python
import pandas as pd
```

Tabla Calendario.

```python
start, end = '1960-01-01', '2025-09-19'

calendario = pd.DataFrame(pd.date_range(start, end), columns=['fecha'])
calendario.head()
```

```python
calendario['dia'] = calendario.fecha.apply(lambda x: x.day)
calendario['mes'] = calendario.fecha.apply(lambda x: x.month)
calendario['anio'] = calendario.fecha.apply(lambda x: x.year)
calendario['diaSemana'] = calendario.fecha.apply(lambda x: x.day_name())
calendario['semana'] = calendario.fecha.apply(lambda x: x.week)
```

```python
map_dias = {
    "Monday": 0,
    "Tuesday": 1,
    "Wednesday": 2,
    "Thursday": 3,
    "Friday": 4,
    "Saturday": 5,
    "Sunday": 6
}

calendario['diaNumeroSemana'] = calendario.diaSemana.map(map_dias)
calendario.sample(3)
```

```python
calendario['trimestre'] = calendario.mes.apply(lambda x: 1 if x <= 3 else 2 if x <= 6 else 3 if x <= 9 else 4)
calendario['semestre'] = calendario.mes.apply(lambda x: 1 if x <= 6 else 2)

calendario.sample(3)
```

Quiero convertir los nombres de los dias a espaÃ±ol.

```python
map_dias_es = {
    "Monday": "Lunes",
    "Tuesday": "Martes",
    "Wednesday": "MiÃ©rcoles",
    "Thursday": "Jueves",
    "Friday": "Viernes",
    "Saturday": "SÃ¡bado",
    "Sunday": "Domingo"
}

calendario.diaSemana = calendario.diaSemana.map(map_dias_es)
calendario.sample(3)
```

```python
def es_bisiesto(anio: int) -> bool:
    """
    Devuelve True si el aÃ±o es bisiesto segÃºn el calendario gregoriano.
    """
    return (anio % 4 == 0) and (anio % 100 != 0 or anio % 400 == 0)

calendario['bisiesto'] = calendario.anio.apply(es_bisiesto)
calendario.sample(5)
```

Columna de quincena

```python
calendario['quincena'] = calendario.dia.apply(lambda x: 1 if x <= 15 else 2)
calendario.sample()
```

Columna de nÃºmero de semana en el mes

```python
calendario['semanaMes'] = calendario.dia.apply(lambda x: 1 if x <= 7 else 2 if x <= 14 else 3 if x <= 21 else 4)
calendario.sample()
```

```python
calendario = calendario[[
    'fecha',
    'anio', 'mes', 'dia',
    'trimestre', 'semestre', 'quincena', 'semanaMes', 'semana',
    'diaSemana', 'diaNumeroSemana',
    'bisiesto'
]]

calendario.sample(5)
```

```python
calendario.to_csv('calendario.csv', index=False)
```
