```python
import pandas as pd
import glob
```

```python
datos = "E:/Facux/Datos poblacion/*.csv"

archivos = glob.glob(datos)
poblaciones = pd.concat([pd.read_csv(a) for a in archivos], ignore_index=True)

print(poblaciones.shape)
poblaciones.head()
```

```python
poblaciones.to_csv("poblaciones.csv", index=False, encoding="utf-8")
```
